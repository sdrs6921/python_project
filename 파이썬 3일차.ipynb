```python
from csv import reader

opened_file = open('AppleStore.csv', encoding='UTF8')
read_file = reader(opened_file)
apps_data = list(read_file)
```

```python
content_ratings = {
    '4+': 0,
    '9+': 0,
    '12+': 0,
    '17+': 0
}
```

```python
for data in apps_data[1:]:
    cont_rating = data[10]
    
    if cont_rating in content_ratings:
        content_ratings[cont_rating] += 1
print(content_ratings)
```

```python
#앱 장르별 카운팅
#1. 빈 딕셔너리 생성
#2. 장르의 index 11
#3. if-else문으로 새로운 데이터 입력시 추가하는 제어문
genre_counting = {}

for data in apps_data[1:]:
    genre = data[11]
    
    if genre in genre_counting:
        genre_counting[genre] += 1
    else:
        genre_counting[genre] = 0
    
print(genre_counting)
```

```python
#앱 연령별 percentage 분석
#1.각 요소에 100을 곱하여 퍼센티지로 변환
#2.17+의 percentage를 출력
#3.15세인 사람ㄷ이 다운로드 할 수 잇는 퍼센티지를 출력
total_number_of_apps = 7197

for iterable_variable in content_ratings:
    content_ratings[iterable_variable] /= total_number_of_apps
    content_ratings[iterable_variable] *= 100#1

percentage_17_plus = content_ratings['17+']#2
percentage_15_allowed = content_ratings['4+'] + content_ratings['9+'] + content_ratings['12+']#3

print(percentage_17_plus)
print(percentage_15_allowed)
```

```python
#앱 사이즈의 최대 값, 최소 값 분석
#1. 빈 리스트 생성
#2. 리스트의 최대 최소 값 출력
size_bytes = []

for data in apps_data[1:]:
    size = float(data[2])
    size_bytes.append(size)

print(max(size_bytes))
print(min(size_bytes))
```

```python
#앱 토탈 평점 개수 구간 별 카운팅
#1. rating_count_tot - 5번 index
#2. 구간 설정
#3. 구간 개수 카운팅
#4. 컬럼의 최대 최소값 계산
rating_counts = []
for data in apps_data[1:]:
    rating_count = float(data[5])
    rating_counts.append(rating_count)
print(max(rating_counts))
print(min(rating_counts))
```

```python
ratgins_counting_tot_distribution = {
    '0 - 500000': 0, 
    '500000 - 1000000': 0,
    '1000000 - 1500000': 0,
    '1500000 - 2000000': 0, 
    '2000000 - 2500000': 0,
    '2500000 - 3000000': 0
}

for rating_count_total in rating_counts:
    if rating_count_total >= 0.0 and rating_count_total > 500000.0:
        ratgins_counting_tot_distribution['0 - 500000'] += 1
    elif rating_count_total >= 500000.0 and rating_count_total > 1000000.0:
        ratgins_counting_tot_distribution['500000 - 1000000'] += 1
    elif rating_count_total >= 1000000.0 and rating_count_total > 1500000.0:
        ratgins_counting_tot_distribution['1000000 - 1500000'] += 1
    elif rating_count_total >= 1500000.0 and rating_count_total > 2000000.0:
        ratgins_counting_tot_distribution['1500000 - 2000000'] += 1
    elif rating_count_total >= 2000000.0 and rating_count_total > 2500000.0:
        ratgins_counting_tot_distribution['2000000 - 2500000'] += 1
    else:
        ratgins_counting_tot_distribution['2500000 - 3000000'] += 1
print(ratgins_counting_tot_distribution)
```

```python
"""
앱스토어 데이터 중 특정 컬럼 추출 함수 작성
1. 앱스토어 데이터의 특정 컬럼을 추출하여 리스트로 만들어 리턴
2. 함수 이름: extract
3. 인자 추출 대상 index 번호
4. 리턴값 - 추출 된 컬럼의 리스트
5. 
"""

def extract(index):
    """
    extract AppleStore.csv File column
    parameter : column index (0 ~ 16 integer number)
    return : extracted column list
    """
    opened_file = open('AppleStore.csv', encoding='UTF8')
    read_file = reader(opened_file)
    apps_data = list(read_file)
    
    extracted_column_list = []
    
    for data in apps_data[1:]:
        extracted_column_data = data[index]
        extracted_column_list.append(extracted_column_data)
    
    return extracted_column_list

print(extract(1))
```

```python
def extract(*index):
    """
    extract AppleStore.csv File column
    parameter : column index (0 ~ 16 integer number)
    return : extracted column list
    """
    opened_file = open('AppleStore.csv', encoding='UTF8')
    read_file = reader(opened_file)
    apps_data = list(read_file)
    
    extracted_column_list = []
    
    for data in apps_data[1:]:
        extracted_column_data = []
        for i in index:
            extracted_column_data.append(data[i])
        extracted_column_list.append(extracted_column_data)
    
    return extracted_column_list

print(extract(1, 2))
```

```python
"""
해당 index의 딕셔너리의 각 값이 몇 번 등장했는지. / rating or genre 7, 11
"""
def freq_table(index):
    """
    extract frequency table
    parameter : column index (7 : rating  or 11: genre )
    return : value frequency table dictionary
    """
    opened_file = open('AppleStore.csv', encoding='UTF8')
    read_file = reader(opened_file)
    apps_data = list(read_file)
    
    frequency_table_dict = {}
    
    for data in apps_data[1:]:
        data_key = data[index]
        
        if data_key in frequency_table_dict:
            frequency_table_dict[data_key] += 1
        else:
            frequency_table_dict[data_key] = 1
            
    return frequency_table_dict

print(freq_table(7))
print(freq_table(11))
```
